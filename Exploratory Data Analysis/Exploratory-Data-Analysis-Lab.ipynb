## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here

plt.figure(figsize=(10, 6))
df.boxplot(column='ConvertedComp')
plt.ylabel('ConvertedComp')
plt.title('Box Plot of ConvertedComp')
plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here

Q1 = df['ConvertedComp'].quantile(0.25)
Q3 = df['ConvertedComp'].quantile(0.75)
IQR = Q3 - Q1
print("Inter Quartile Range (IQR) for ConvertedComp:", IQR)
```

Find out the upper and lower bounds.


```python
# your code goes here

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here

outliers = df[(df['ConvertedComp'] < lower_bound) | (df['ConvertedComp'] > upper_bound)]
num_outliers = len(outliers)
print("Number of outliers in ConvertedComp:", num_outliers)
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here

df_no_outliers = df[(df['ConvertedComp'] >= lower_bound) & (df['ConvertedComp'] <= upper_bound)]
```

```python
# the new median ConvertedComp after removing outliers
new_median_converted_comp = df_no_outliers['ConvertedComp'].median()

print("New Median ConvertedComp (after removing outliers):", new_median_converted_comp)
```

```python
# the new mean ConvertedComp after removing outliers
new_mean_converted_comp = df_no_outliers['ConvertedComp'].mean()

print("New Mean ConvertedComp (after removing outliers):", new_mean_converted_comp)
```

## Correlation


### Finding correlation


Find the correlation between `Age` and all other numerical columns.


```python
# Select the numerical columns in the dataframe
numerical_columns = df.select_dtypes(include='number')

# Calculate the correlation between Age and all other numerical columns
correlation = numerical_columns.corr()['Age']

print(correlation)


#we select only the numerical columns using the select_dtypes(include='number') method.
#Next, we calculate the correlation between the "Age" column and all other numerical columns using the .corr() method. 
#The result is stored in the correlation variable.
```
